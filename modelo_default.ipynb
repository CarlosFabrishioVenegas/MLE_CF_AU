```python
import pandas as pd
from pycaret.classification import *
```

```python
#1. Cargamos la data
df = pd.read_csv('ds_credito_vehicular.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe().T
```

```python
clf = setup(
    data = df,
    target = 'default',
    session_id = 123,
    categorical_features = ['estado_civil','historia_credito','vehiculo_propio'],
    normalize=True,
    fix_imbalance=True,
    fix_imbalance_method = 'SMOTE'
)
```

```python
best_model = compare_models()
```

```python
lr = create_model("lr")
```

```python
tunned_lr = tune_model(lr)
```

```python
#Evaluarlo visualmente
evaluate_model(tunned_lr)
```

```python
preds = predict_model(tunned_lr)
print(preds.head())
```

```python
# Guardar el modelo
save_model(tunned_lr, "modelo_default_final")
```
